Q.2. In which year were the highest number of TV Shows and movies released? Visualize using bar graph. 

```
df.dtypes
```

```
# Add a new column to the dataframe which shows year of release
df['Year'] = pd.to_datetime(df['Release_Date']).dt.year.astype('Int64')
```

```
df.head()
```

```
df.dtypes
```

```
# Counts the occurence of all individual years in date column
df['Year'].value_counts().sort_index()
```

Bar Graph

```
from matplotlib import pyplot as plt
import numpy as np

x = df['Year'].value_counts().sort_index().index 
y = df['Year'].value_counts().sort_index().values

fig, ax = plt.subplots()
fig.set_figwidth(11)
bar_container = ax.barh(x,y,color='royalblue', )
ax.set(yticks=np.arange(x[0],x[-1]+1,1),xlabel='Number of shows', ylabel='Year',title='Number of Movies and TV Shows Released Each Year')
ax.bar_label(bar_container, labels=[f'{x:,.0f}' for x in bar_container.datavalues])
plt.show()
```

Q.3. How many movies and TV shows are in the dataset? Show with bar graph.

```
df.groupby('Category').Category.count()
```

```
x = df.groupby('Category').Category.count().values
y = df.groupby('Category').Category.count().index 
fig, ax = plt.subplots()
bar_container = ax.bar(y,x,color=['cornflowerblue','orchid'] )
ax.set(xlabel='Category', ylabel='Count',title='Total number of movies and TV shows in the dataset')
ax.bar_label(bar_container, labels=[f'{x:,.0f}' for x in bar_container.datavalues])
plt.show()
```
